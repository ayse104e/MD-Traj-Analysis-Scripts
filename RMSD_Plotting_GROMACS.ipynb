```python
import numpy as np

# Load data from the xvg file
data = np.loadtxt('rmsd_ligand.xvg', comments=['@', '#'])
time = data[:, 0]  # Time in ps
rmsd_values = data[:, 1]  # RMSD values in nm
```

```python
import matplotlib.pyplot as plt

# Plot RMSD values
plt.figure(figsize=(10, 6))
plt.plot(time, rmsd_values, linewidth=2)
plt.title('RMSD of Ligand Atoms')
plt.xlabel('Time (ps)')
plt.ylabel('RMSD (nm)')
plt.grid(True)
plt.tight_layout()

# Save or Show the plot
plt.savefig('rmsd_ligand_plot.png', dpi=300)  # Save plot as PNG
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load data from the xvg file
data = np.loadtxt('rmsd_ligand.xvg', comments=['@', '#'])
time = data[:, 0]  # Time in ps
rmsd_values = data[:, 1]  # RMSD values in nm

# Calculate mean RMSD and standard deviation
mean_rmsd = np.mean(rmsd_values)
std_dev = np.std(rmsd_values)

# Plot RMSD values
plt.figure(figsize=(10, 6))
plt.plot(time, rmsd_values, linewidth=2, label='RMSD of Ligand')
plt.axhline(mean_rmsd, color='red', linestyle='--', label=f'Mean RMSD: {mean_rmsd:.3f} nm\nStandard Deviation: {std_dev:.3f} nm')
plt.title('RMSD of Ligand Atoms')
plt.xlabel('Time (ps)')
plt.ylabel('RMSD (nm)')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Save or Show the plot
plt.savefig('rmsd_ligand_plot_with_mean_std.png', dpi=300)  # Save plot as PNG
plt.show()
```
